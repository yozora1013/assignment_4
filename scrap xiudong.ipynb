```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

header = ["style",'title',"link",'artist','time',"price","address"]
data = []
title = []
artist=[]
price = []
time = []
addr=[]
link=[]
style=[1,2,3,4,6,23,26]

for i in style: 
    url = "https://www.showstart.com/event/list?pageNo=1&pageSize=20&cityCode=10&showTime=6&sortType=9&showStyle="+str(i)
    quote_page = requests.get(url) 
    soup = BeautifulSoup(quote_page.content,'html.parser')
    
    for item in soup.find_all("a", attrs={"class":"show-item item"}):
         style_no = str(i)
         title = item.find("div",attrs={"class":"title"}).get_text()
         link = item.get("href")
         artist = item.find("div",attrs={"class":"artist"}).get_text()
         time = item.find("div",attrs={"class":"time"}).get_text()
         price = item.find("div",attrs={"class":"price"}).get_text()
         address = item.find("div",attrs={"class":"addr"}).get_text()
         data.append((style_no,title, link,artist,time,price,address))
        
df = pd.DataFrame(data, columns = header)
df.to_csv('liveshow_timetable.csv')
df
```

```python
import pandas as pd
df = pd.read_csv("liveshow_timetable.csv")

df["is_duplicate"] = df.duplicated("link")
df_removed=df.loc[0:53,'style':'address']
df_removed
df = df_removed.drop_duplicates("link")

df["style"].replace(2,"摇滚", inplace=True)
df["style"].replace(26,"独立", inplace=True)
df["style"].replace(23,"朋克", inplace=True)
df["style"].replace(3,"流行", inplace=True)
df["style"].replace(4,"电子", inplace=True)
df["style"].replace(6,"HipHop", inplace=True)
df["style"].replace(1,"民谣", inplace=True)
df["artist"]= df["artist"].str.replace("艺人：","")
df["time"]= df["time"].str.replace("时间：","")
df["price"]= df["price"].str.replace("价格：","")
df["address"]= df["address"].str.replace("北京]","")
df["address"]= df["address"].str.replace("[","")
df["link"]= df["link"].str.replace('/event', 'https://www.showstart.com/event')

df
```
